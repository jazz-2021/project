```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
api_key = "4993d963bc4cad9a9686b6dbf2ed08a8"
start_date = "2015-01-01"
end_date = "2015-12-31"


forex_query = f"https://api.metalpriceapi.com/v1/timeframe?api_key={api_key}&start_date={start_date}&end_date={end_date}&base=USD&currencies=MYR"
gold_query = f"https://api.metalpriceapi.com/v1/timeframe?api_key={api_key}&start_date={start_date}&end_date={end_date}&base=USD&currencies=XAU"
silver_query = f"https://api.metalpriceapi.com/v1/timeframe?api_key={api_key}&start_date={start_date}&end_date={end_date}&base=USD&currencies=XAG"


forex_record = requests.get(forex_query)
gold_record = requests.get(gold_query)
silver_record = requests.get(silver_query)
silver_record
```

```python
forex_json = forex_record.json()
gold_json = gold_record.json()
silver_json = silver_record.json()

dates = list(forex_json["rates"].keys())
forex_rates = [forex_json["rates"][val]["MYR"] for val in dates]
gold_rates = [gold_json["rates"][val]["USDXAU"] for val in dates]
silver_rates = [silver_json["rates"][val]["USDXAG"] for val in dates]

historical_df = pd.DataFrame(list(zip(dates, forex_rates, gold_rates, silver_rates)), 
                             columns=["date", "usd_to_myr", "gold_troy_oz_usd", "silver_troy_oz_usd"])
historical_df
```

```python
historical_df.to_csv("precious_metal_2015.csv", index=False)
```
